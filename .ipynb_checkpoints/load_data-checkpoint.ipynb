```python
import pymongo
import pandas as pd
import json
```

```python
conn = "mongodb://localhost:27017"
client = pymongo.MongoClient(conn)
```

```python
db = client.lawn_mower_inventory
collection = db.lawn_mower
```

```python
mower_data_cleaned = "mower_data_cleaned.csv"
jonathan_df = pd.read_csv(mower_data_cleaned)
jonathan_df
```

```python
jonathan_json = jonathan_df.to_json(orient='records')
jonathan_json
```

```python
parsed = json.loads(jonathan_json)
print(json.dumps(parsed, indent=4, sort_keys=True))
```

```python
db.collection.insert_many(jonathan_json)
```
